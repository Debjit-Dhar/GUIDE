```python
# DiCMA Analysis Notebook

This notebook provides a lightweight workspace for exploring the DiCMA losses and verifying behavior.

**Sections:**
1. Load previous state / imports
2. Inspect variables and outputs
3. Continue data processing / experiments
4. Save progress
</VSCode.Cell>
<VSCode.Cell language="markdown">
## 1) Load Previous Notebook State

If you have any saved checkpoint or intermediate results, load them here. For this minimal example we will just import the DiCMA module and create a small toy dataset.
</VSCode.Cell>
<VSCode.Cell language="python">
# Imports
import torch
from dicma.gaussian_prototypes import GaussianPrototypes
from dicma.losses import w2_gaussian_squared, covariance_frobenius_loss

print("PyTorch version:", torch.__version__)

# Create a toy DiCMA module
num_ids = 4
feat_dim = 32
proto = GaussianPrototypes(num_ids=num_ids, feat_dim=feat_dim, rank=16)

# Create toy batch data
batch_size = 12
features = torch.randn(batch_size, feat_dim)
ids = torch.randint(high=num_ids, size=(batch_size,))

out = proto(features, ids)
print("W2 loss:", out["w2_loss"].item())
print("Cov loss:", out["cov_loss"].item())
</VSCode.Cell>
<VSCode.Cell language="markdown">
## 2) Inspect Variables and Outputs

Check the current losses and statistics, and inspect per-ID moment estimates.
</VSCode.Cell>
<VSCode.Cell language="python">
# Examine prototype means and a few running statistics
for idx in range(min(3, num_ids)):
    mu, cov = proto.get_prototype(idx)
    print(f"ID {idx}: prototype mean norm={mu.norm().item():.4f}, cov trace={torch.trace(cov).item():.4f}")

# Print running means for first few IDs
print("Running means (first 3 IDs):")
print(proto.running_mean[:3])
</VSCode.Cell>
<VSCode.Cell language="markdown">
## 3) Continue Data Processing

You can extend this section to load real features from your model, compute DiCMA losses during training, and log diagnostics (W2 values, covariance mismatch, margins).
</VSCode.Cell>
<VSCode.Cell language="python">
# Example: update running stats with another batch
features2 = torch.randn(batch_size, feat_dim)
ids2 = torch.randint(high=num_ids, size=(batch_size,))
out2 = proto(features2, ids2)
print("Updated W2 loss:", out2["w2_loss"].item())
</VSCode.Cell>
<VSCode.Cell language="markdown">
## 4) Save Progress

You can save the module state, or export metrics to disk.
</VSCode.Cell>
<VSCode.Cell language="python">
# Save prototype weights for later inspection
torch.save(proto.state_dict(), "dicma/prototype_state.pth")
print("Saved prototype state to dicma/prototype_state.pth")
</VSCode.Cell>
```
